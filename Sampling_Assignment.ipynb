```python
def systematic_sampling(df, step):
 
    indexes = np.arange(0, len(df), step=step)
    systematic_sample = df.iloc[indexes]
    return systematic_sample
systematic_sample = systematic_sampling(data,3)
```

```python
def get_clustered_Sample(df, n_per_cluster, num_select_clusters):
    N = len(df)
    K = int(N/n_per_cluster)
    data = None
    for k in range(K):
        sample_k = df.sample(n_per_cluster)
        sample_k["cluster"] = np.repeat(k,len(sample_k))
        df = df.drop(index = sample_k.index)
        data = pd.concat([data,sample_k],axis = 0)

    random_chosen_clusters = np.random.randint(0,K,size = num_select_clusters)
    samples = data[data.cluster.isin(random_chosen_clusters)]
    return(samples)

cluster_sample = get_clustered_Sample(df = data, n_per_cluster = 150, num_select_clusters = 3)
cluster_sample
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC 
```

```python
def model_training(df):
    clf1 = LogisticRegression(max_iter=1000, verbose=False)
    clf2  = LinearSVC()
    clf3 = RandomForestClassifier()
    clf4 = DecisionTreeClassifier()
    clf5 = SVC()
    X = df.iloc[:,0:ncols-1]
    Y = df.iloc[:,ncols-1]
    X = np.array(X)
    Y = np.array(Y)
    clf1.fit(X,Y)
    clf2.fit(X,Y)
    clf3.fit(X,Y)
    clf4.fit(X,Y)
    clf5.fit(X,Y)
    models = [clf1, clf2, clf3, clf4,clf5]
    return models
```

```python
from sklearn.metrics import accuracy_score
```

```python
def accuracy_score_calc(clf_list, X_test, Y_test):
    for i in range(len(clf_list)):
        model = clf_list[i]
        Y_Predict = model.predict(X_test)
        score = accuracy_score(y_true = Y_test, y_pred = Y_Predict)
        print("Accuracy Score for model: ",i+1, "=",score)
```

```python
for i in range(len(df_list)):
    curr_df = df_list[i]
    model_list = model_training(curr_df)
    accuracy_score_calc(model_list, X_test, Y_test)
```
